```python
import numpy as np
import matplotlib.pyplot as plt
import random

N = 1000
repeat = 10
x_start, y_start = 0, 0
x, y = [0], [0]

plt.figure(figsize=(5, 5))
plt.xlim(-40, 40)
plt.ylim(-40, 40)
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.title("Random walk")

for i in range(repeat):
    for j in range(N):
        x_start += (random.random() - 0.5) * 2
        y_start += (random.random() - 0.5) * 2
        x.append(x_start)
        y.append(y_start)
    plt.plot(x, y, linewidth=0.5)
    x_start, y_start = 0, 0
    x, y = [0], [0]
    
plt.savefig("walk.jpg", dpi=200)
```
