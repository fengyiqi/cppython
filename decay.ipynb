```python
import matplotlib.pyplot as plt
import numpy as np

lambda_ = 0.01
particles = 200
end_time = 500
number = particles
res = []
for i in range(end_time):
    decay = np.random.rand(number)
    gone = (decay < lambda_).sum()
    number -= gone
    res.append(number)

plt.figure(figsize=(5, 4))
plt.plot(range(end_time), res, "b-", linewidth=0.8)
plt.xlabel("$t$")
plt.ylabel("Number left")
plt.title("Spontaneous Decay")
plt.yscale("log")
plt.savefig("decay.jpg", dpi=200)
```
